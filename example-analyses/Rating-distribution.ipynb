```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_dist_plot(contest, alg="KLUCB"):
    df = pd.read_csv(f'../contests/summaries/{contest}_summary_{alg}.csv')

    ax = df.score.hist(bins=20, grid=False)
    ax.set_xlim(1, 3)
    ax.set_xticks([1, 2, 3])
    ax.set_xticklabels(["Unfunny", "Somewhat funny", "Funny"])
    ax.set_title(f"Ratings for contest={contest}")
    return ax
```

```python
get_dist_plot(538, alg="LilUCB")
```

```python
get_dist_plot(640)
```
